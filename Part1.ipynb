```python
import json
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import nltk
import gzip
from collections import Counter
```

# 1.2 Load Dataset

```python
f = gzip.open('goemotions.json.gz', 'rb')
data = json.load(f)
```

# 1.3 Extract posts (both emotion and sentiment) and Plot Distribution in Pie Chart
## Emotion

```python
c = Counter(comment[1] for comment in data)

emotionalPie = plt.pie( c.values(),labels=c.keys(), autopct='%1.1f%%', radius=5, shadow=True)
plt.savefig('emotions.png', bbox_inches = 'tight')
plt.show()
```

## Sentiment

```python
c = Counter(comment[2] for comment in data)

sentimentPie = plt.pie( c.values(),labels=c.keys(), autopct='%1.1f%%', radius=5)
plt.savefig('sentiment.png')
plt.show()
```
